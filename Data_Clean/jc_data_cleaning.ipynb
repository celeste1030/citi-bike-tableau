```python
import pandas as pd
```

```python
file_path2 = "Data/jc_202012.csv"

jc_df = pd.read_csv(file_path2)

jc_df.head()
```

```python
jc_df["startdate"] = jc_df.starttime.str.split(expand = True)[0]
jc_df["enddate"] = jc_df.stoptime.str.split(expand = True)[0]

jc_df.head()
```

```python
jc_df.drop(inplace= True, columns = ["starttime", "stoptime"])

jc_df.head()
```

```python
jc_df.replace(inplace= True, to_replace={0: 'undisclosed', 1: 'male', 2: 'female'})

jc_df.head()
```

```python
#jc_df.sort_values(by = 'birth year', ascending=True)

years = jc_df['birth year'].value_counts().index.tolist()
years.sort()
print(years)
```

```python
bins = [1940, 1950, 1960, 1970, 1980, 1990, 1999, 2004]

group_labels = ["70-79", "60-69", "50-59", "40-49", "30-39", "21-29", "Under 21"]

pd.cut(jc_df["birth year"], bins, labels=group_labels).head()
```

```python
jc_df["age"] = pd.cut(jc_df["birth year"], bins, labels=group_labels)
jc_df.head()
```

```python
jc_trips = pd.DataFrame(jc_df.groupby(['startdate', 'usertype', 'gender']).agg(
    min_trip=pd.NamedAgg(column="tripduration", aggfunc="min"),
    max_trip=pd.NamedAgg(column="tripduration", aggfunc="max"),
    avg_trip=pd.NamedAgg(column="tripduration", aggfunc="mean"))).reset_index()
        
jc_trips
```

```python
jc_trips.to_csv("Output/jc_trips.csv", encoding='utf-8', index=False)
```

```python
sub_count = jc_df["usertype"].value_counts().to_frame().reset_index()
sub_count.rename(inplace= True, columns= {'index': 'usertype', 'usertype': 'count'})

sub_count.to_csv("Output/jc_subcount.csv", encoding='utf-8', index=False)
```

```python
data = jc_df.copy()

stations = pd.get_dummies(data, columns=["usertype"], dtype=int)
stations
```

```python
jc_start = stations.drop(columns=['end station id', 'end station name',
                         'end station latitude', 'end station longitude', 'bikeid', 'birth year', 'gender', 'enddate','age', 'usertype_Customer'], axis= 1)
jc_start
```

```python
jc_stat_start = pd.DataFrame(jc_start.groupby(['start station id', 'usertype_Subscriber']).agg(
    avg_trip=pd.NamedAgg(column="tripduration", aggfunc="mean"),
    station_name=pd.NamedAgg(column="start station name", aggfunc="min"),
    station_lat=pd.NamedAgg(column="start station latitude", aggfunc="max"),
    station_long=pd.NamedAgg(column="start station longitude", aggfunc="max"),
    rides=pd.NamedAgg(column="startdate", aggfunc="count"))).reset_index()

jc_stat_start.rename(inplace= True, columns={"usertype_Subscriber":"subscriber"})
#subscribers=pd.NamedAgg(column="usertype_Subscriber", aggfunc="count"),


jc_stat_start
```

```python
jc_stat_start.to_csv("Output/jc_stations_start.csv", encoding='utf-8', index=True)
```

```python
jc_end = stations.drop(columns=['start station id', 'start station name', 'start station latitude', 'start station longitude', 'bikeid', 'birth year', 'gender', 'age', 'usertype_Customer'], axis= 1)
jc_end
```

```python
jc_stat_end = pd.DataFrame(jc_end.groupby(['end station id', 'usertype_Subscriber']).agg(
    avg_trip=pd.NamedAgg(column="tripduration", aggfunc="mean"),
    station_name=pd.NamedAgg(column="end station name", aggfunc="min"),
    station_lat=pd.NamedAgg(column="end station latitude", aggfunc="max"),
    station_long=pd.NamedAgg(column="end station longitude", aggfunc="max"),
    rides=pd.NamedAgg(column="enddate", aggfunc="count"))).reset_index()

jc_stat_end.rename(inplace= True, columns={"usertype_Subscriber":"subscriber"})
#subscribers=pd.NamedAgg(column="usertype_Subscriber", aggfunc="count"),


jc_stat_end
```

```python
jc_stat_end.to_csv("Output/jc_stations_end.csv", encoding='utf-8', index=True)
```
