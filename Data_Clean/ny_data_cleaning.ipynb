```python
import pandas as pd
```

```python

file_path = "Data/nyc_202012.csv"

nyc_df = pd.read_csv(file_path)

nyc_df.head()
```

```python
nyc_df["startdate"] = nyc_df.starttime.str.split(expand = True)[0]
nyc_df["enddate"] = nyc_df.stoptime.str.split(expand = True)[0]

nyc_df.head()
```

```python
nyc_df.drop(inplace= True, columns = ["starttime", "stoptime"])

nyc_df.head()
```

```python
nyc_df.replace(inplace= True, to_replace={0: 'undisclosed', 1: 'male', 2: 'female'})

nyc_df.head()
```

```python
#nyc_df.sort_values(by = 'birth year', ascending=True)

years = nyc_df['birth year'].value_counts().index.tolist()
years.sort()
print(years)
```

```python
#find oldest and youngest year
#oldest = nyc_df['birth year'].min()
#youngest = nyc_df['birth year'].max()

#bins

bins = [1885, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 1999, 2004]

group_labels = ["null", "90-94", "80-89", "70-79", "60-69", "50-59", "40-49", "30-39", "21-29", "Under 21"]

pd.cut(nyc_df["birth year"], bins, labels=group_labels).head()
```

```python
nyc_df["age"] = pd.cut(nyc_df["birth year"], bins, labels=group_labels)
nyc_df.head()
```

```python
#nyc_df.drop(nyc_df.loc[nyc_df['age']=='null'].index, inplace=True)
#nyc_df
```

```python
nyc_trips = pd.DataFrame(nyc_df.groupby(['startdate', 'usertype', 'gender']).agg(
    min_trip=pd.NamedAgg(column="tripduration", aggfunc="min"),
    max_trip=pd.NamedAgg(column="tripduration", aggfunc="max"),
    avg_trip=pd.NamedAgg(column="tripduration", aggfunc="mean"))).reset_index()
        
nyc_trips
```

```python
nyc_trips.to_csv("Output/nyc_trips.csv", encoding='utf-8', index=False)
```

```python
sub_count = nyc_df["usertype"].value_counts().to_frame().reset_index()
sub_count.rename(inplace= True, columns= {'index': 'usertype', 'usertype': 'count'})

sub_count

sub_count.to_csv("Output/nyc_subcount.csv", encoding='utf-8', index=False)
```

```python
user_age = pd.DataFrame(nyc_df.groupby('age')['usertype'].value_counts())
user_age.rename(inplace= True, columns= {'age': 'age', 'usertype': 'usertype', 'usertype': 'count'})

user_age.to_csv("Output/nyc_user_age.csv", encoding='utf-8', index=True)
user_age
```

```python
data = nyc_df.copy()

stations = pd.get_dummies(data, columns=["usertype"], dtype=int)
stations.head()
```

```python
nyc_start = stations.drop(columns=['end station id', 'end station name',
                         'end station latitude', 'end station longitude', 'bikeid', 'birth year', 'gender', 'enddate','age', 'usertype_Customer'], axis= 1)
nyc_start
```

```python
#customers = nyc_start.loc[nyc_start["usertype"] == "Customer"]
#customers=pd.NamedAgg(nyc_start.loc[nyc_start["usertype"] == "Customer"].count(),
#subscribers=pd.NamedAgg(nyc_start.loc[nyc_start["usertype"] == "Subscriber"].count(),

nyc_stat_start = pd.DataFrame(nyc_start.groupby(['start station id', 'usertype_Subscriber']).agg(
    avg_trip=pd.NamedAgg(column="tripduration", aggfunc="mean"),
    station_name=pd.NamedAgg(column="start station name", aggfunc="min"),
    station_lat=pd.NamedAgg(column="start station latitude", aggfunc="max"),
    station_long=pd.NamedAgg(column="start station longitude", aggfunc="max"),
    rides=pd.NamedAgg(column="startdate", aggfunc="count"))).reset_index()


nyc_stat_start.rename(inplace= True, columns={"usertype_Subscriber":"subscriber"})
stations

nyc_stat_start
```

```python
nyc_stat_start.to_csv("Output/nyc_stations_start.csv", encoding='utf-8', index=True)
```
